```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# tuodaan pingviiniaineisto:
penguins = sns.load_dataset('penguins')
penguins
```

```python
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots(2,3, figsize=(10,6))
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].scatter(data=penguins, x='bill_length_mm', y='flipper_length_mm')
ax[0].set_xlabel('Nokan pituus (mm)')
ax[0].set_ylabel('Siiven pituus (mm)')
ax[1].hist(data=penguins, x='body_mass_g')
ax[1].set_xlabel('Massa (g)')
ax[1].set_ylabel('LukumÃ¤Ã¤rÃ¤')
fig.suptitle('Tietoja pingviineistÃ¤')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(2,2, figsize=(9,9), sharex='row')
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species', ax=ax[0,0])
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='sex', ax=ax[0,1])
sns.histplot(data=penguins[penguins['sex']=='Male'], x='body_mass_g', ax=ax[1,0])
sns.histplot(data=penguins[penguins['sex']=='Female'], x='body_mass_g', ax=ax[1,1])
ax[0,0].set_xlabel('Nokan pituus (mm)')
ax[0,0].set_ylabel('Siiven pituus (mm)')
ax[0,1].set_xlabel('Nokan pituus (mm)')
ax[0,1].set_ylabel('Siiven pituus (mm)')
ax[1,0].set_xlabel('Massa (g)')
ax[1,0].set_ylabel('LukumÃ¤Ã¤rÃ¤')
ax[1,1].set_xlabel('Massa (g)')
ax[1,1].set_ylabel('LukumÃ¤Ã¤rÃ¤')
ax[0,0].set_title('Pingviinien nokan ja siiven pituus lajeittain')
ax[0,1].set_title('Pingviinien nokan ja siiven pituus sukupuolittain')
ax[1,0].set_title('Koiraspingviinien painojakauma')
ax[1,1].set_title('Naaraspingviinien painojakauma')
fig.tight_layout(w_pad=5, h_pad=2)
fig.savefig('pingviinikuva.png', bbox_inches='tight')
```
